## Questão 5  📝

Peça ao usuário para digitar uma nota (0 a 10) e classifique usando `if`/`elif`/`else`:

- Nota >= 7: `"Aprovado"`
- Nota >= 5: `"Recuperação"`
- Nota < 5: `"Reprovado"`

````python
nota = float(input("Nota: "))

if nota >= 0 and nota <= 10:

    if nota >= 7:
        print("Aprovado")
        
    elif nota >= 5:
        print("Recuperação")

    else: 
        print("Reprovação")

else:
    print("Essa nota não existe!")
````

### 💡 Dica:

*Atenção à ordem das condições, se entrar no primeiro if, o resto das condicionais será pulado. Comece sempre pelo maior valor.*

## Questão 6 📝

Crie um programa que recebe o salário de um funcionário e calcula o imposto devido com base nas faixas abaixo:

| Faixa salarial | Alíquota |
|---|---|
| Até R\$ 1.500 | Isento (0%) |
| R\$ 1.500,01 – R\$ 3.000 | 7,5% |
| R\$ 3.000,01 – R\$ 4.500 | 15% |
| Acima de R\$ 4.500 | 22,5% |

Exiba o salário bruto, o valor do imposto e o salário líquido.

````python
salario = float(input("Salário: "))

if salario > 0:

    if salario <= 1500:

        print("Salário bruto: ", salario)
        print("Imposto: Isento(%0)")
        print("Salário líquido: ", salario)

    elif salario > 1500 and salario <= 3000: 

        imposto = (salario - 1500)*0.075

        print("Salário bruto: ", salario)
        print("Imposto: ", imposto)
        print("Salário líquido: ", salario - imposto)

    elif salario > 3000 and salario <= 4500:

        imposto = (1500*0.075) + ((salario - 3000)*0.15)

        print("Salário bruto: ", salario)
        print("Imposto: ", imposto)
        print("Salário líquido: ", salario - imposto)

    else:

        imposto = (1500*0.075) + (1500*0.15) + ((salario - 4500)*0.225)

        print("Salário bruto: ", salario)
        print("Imposto: ", imposto)
        print("Salário líquido: ", salario - imposto)     

else:
    print("Valor inválido!")
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Use `if`/`elif`/`else` para cada faixa. Calcule `imposto = salario * aliquota`, depois `liquido = salario - imposto`. *

## Questão 7 📝

Crie um programa que verifica se um ano digitado pelo usuário é bissexto.

**Regras:**
- Divisível por 4 → bissexto
- Mas se divisível por 100 → **NÃO** é bissexto
- A menos que também seja divisível por 400 → **É** bissexto

Exiba `"Ano bissexto"` ou `"Não é bissexto"`.

````python
ano = int(input("Ano: "))

if ano > 0:
        
    if (ano % 4 == 0 and not ano % 100 == 0) or (ano % 400 == 0):
        print("É bissexto")

    else:
        print("Não é bissexto")
else: 
    print("Esse número é inválido")
    
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Use o operador `%` para verificar divisibilidade. Combine condições com `and` e `or`.*

## Questão 8 📝

Construa um programa que recebe o peso e a altura do usuário, calcula o IMC e classifica o resultado:

| IMC | Classificação |
|---|---|
| IMC < 18,5 | Abaixo do peso |
| 18,5 <= IMC < 25 | Peso normal |
| 25 <= IMC < 30 | Sobrepeso |
| IMC >= 30 | Obesidade |

Exiba o IMC com 2 casas decimais e a classificação.

````python
peso = float(input("Peso: "))
altura = float(input("Altura: "))

if (peso > 0 and altura > 0):

    imc = peso/(altura*altura)

    if (imc < 18.5):
        print("IMC: ", f"{imc:.2f}")
        print("Abaixo do peso")

    elif (18.5 <= imc and imc < 25):
        print("IMC: ", f"{imc:.2f}")
        print("Peso normal")

    elif (25 <= imc and imc < 30):
        print("IMC: ", f"{imc:.2f}")
        print("Sobrepeso")
    else:
        print("IMC: ", f"{imc:.2f}")
        print("Obesidade")

else:
    print("Desculpe, não é possível calcular o IMC com esses valores")
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*`IMC = peso / altura**2`. Para formatar com 2 casas decimais, use `f"{imc:.2f}"`.*

## Questão 10 📝

Crie um classificador de triângulos. Receba os 3 lados pelo `input()` e:

1. Verifique se os lados formam um triângulo válido *(a soma de dois lados quaisquer deve ser maior que o terceiro)*
2. Se válido, classifique:
   - **Equilátero:** todos os lados iguais
   - **Isósceles:** dois lados iguais
   - **Escaleno:** todos diferentes

Exiba a classificação ou `"Não forma um triângulo"`.

````python
a = float(input("Primeiro lado: "))
b = float(input("Segundo lado: "))
c = float(input("Terceiro lado: "))

if a + b > c and a + c > b and b + c > a:

    if a == b and a == b and b == c:
        print("Os lados formam um triângulo equilátero")

    elif a == b or a == c or b == c:
        print("Os lados formam um triângulo isósceles!")

    else:
        print("Os lados formam um triângulo escaleno")

else:
    print("Os lados não formam um triângulo")
    

````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Condição de triângulo válido: `a+b>c AND a+c>b AND b+c>a`. Depois compare os lados com `==` para classificar.*
